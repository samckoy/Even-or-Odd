# Even or Odd

### In this program, you will be asked to type in a number, and you will be told whether that number is odd or even.

In the first line of code, the user is asked to enter a number. It must be a number like '14', not 'fourteen'. This number is stored in variable entitled _number_. 

```python
number = int(input("Enter a number: "))
```

A function important to this program is the **modulo** function, or **%**. This function gives you the remainder after dividing two numbers

```python
mod = number % 2
```

All even numbers are divisible by 2. Therefore, if the number you inputed into the program is an even number, that number divided by 2 should have no remainder, and therefore _**number % 2**_ should equal **0**. 

If the number when divided by 2 has a remainder (_mod > 0_), then the number must not be even, and therefore odd. 

```python
if mod > 0:
    print("This number is an odd number.")
else:
    print("This number is an even number.")
```
